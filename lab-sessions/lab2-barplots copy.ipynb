##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
# Importation des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt

# Chargement des données
url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
diamonds = pd.read_csv(url)

# Affichage des 5 premières lignes
print("Premières 5 lignes :")
display(diamonds.head())

# Affichage des noms des colonnes et leurs types
print("\nInformations sur les colonnes :")
print(diamonds.dtypes)

# Affichage des 10 premières lignes
print("\n10 premières lignes :")
display(diamonds.head(10))
```

```python
# Diagramme à barres avec Matplotlib
plt.figure(figsize=(10, 6))
diamonds['cut'].value_counts().plot(kind='bar')
plt.title('Nombre de diamants par qualité de taille')
plt.xlabel('Qualité de taille')
plt.ylabel('Nombre de diamants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Matplotlib : Bibliothèque de base avec une syntaxe impérative, offrant un contrôle total sur tous les aspects du graphique mais nécessitant plus de code. Idéal pour des visualisations personnalisées complexes.


```python
# Diagramme à barres avec Seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=diamonds, x='cut')
plt.title('Nombre de diamants par qualité de taille')
plt.xlabel('Qualité de taille')
plt.ylabel('Nombre de diamants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Seaborn : Extension de Matplotlib avec une syntaxe plus concise et un style par défaut plus moderne. Excellente pour des visualisations statistiques rapides et élégantes.


```python
# Importation des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt

# Chargement des données
url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
diamonds = pd.read_csv(url)

# Création du graphique avec Plotly
df_count = diamonds['cut'].value_counts().reset_index(name='count')
df_count.columns = ['cut', 'count']  # Renommons les colonnes explicitement

fig = px.bar(df_count,
             x='cut',
             y='count',
             title='Nombre de diamants par qualité de taille',
             labels={'cut': 'Qualité de taille', 'count': 'Nombre de diamants'})
fig.show()
```

Plotly : Se distingue par son interactivité native (zoom, survol, déplacement) et sa syntaxe orientée objet. Parfait pour des visualisations web interactives.


```python
# Importation des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt

# Chargement des données
url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
diamonds = pd.read_csv(url)

# Création d'un sous-ensemble des données (par exemple, les 5000 premiers diamants)
diamonds_subset = diamonds.head(5000)

# Création du graphique avec Altair
chart = alt.Chart(diamonds_subset).mark_bar().encode(
    x='cut',
    y='count()'
).properties(
    title='Nombre de diamants par qualité de taille',
    width=600,
    height=400
)
chart
```
