##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt
import pandas as pd
```

```python
#On importe le dataset et on affiche les 5 premiéres lignes
dataset = "./diamonds.xls"
diamonds_df = pd.read_csv(dataset)
diamonds_df.head()
```

```python
#On affiche le type des données
print(diamonds_df.dtypes)
```

```python
#les dix premières lignes
diamonds_df.head(10)
```

```python
#Visualisation avec matplotlib
cut_counts = diamonds_df['cut'].value_counts() # compter les valeurs

plt.figure(figsize=(8, 6))
cut_counts.plot(kind='bar', color='orange')

# Titre + etiquettes
plt.xlabel('Catégorie')
plt.ylabel('Nombre de diamants')
plt.title('Nombre de diamants pour chaque catégorie de cut')
plt.xticks(rotation=45, ha='right')

plt.show

#La syntaxe de Matplotlib est flexible et permet de personnaliser le graphique, comme les couleurs, les titres et les axes etc. 
```

```python
# Visualisation avec seaborn
plt.figure(figsize=(8, 6))
sns.countplot(data= diamonds_df, x='cut', palette='Blues')

# Titre + etiquettes
plt.title('Nombre de diamants pour chaque catégorie de cut', fontsize=14)
plt.xlabel('Catégorie', fontsize=12)
plt.ylabel('Nombre de diamants', fontsize=12)

# Afficher le graphique
plt.show()

# Seaborn simplifie la syntaxe par exemple on a pas besoin de faire une fonction count pour le nombre de diamants comme on a fait avec matplotlib.
```

```python
# Visualisation avec plotly
fig = px.bar(diamonds_df, x="cut", title="Nombre de diamants pour chaque catégorie de cut", labels={'cut':'Catégorie', 'count': 'Nombre de diamonts'}, color='cut')
# Améliorations supplémentaires
fig.update_layout(yaxis_title="Nombre de diamants", showlegend= True)

fig.show()
```
